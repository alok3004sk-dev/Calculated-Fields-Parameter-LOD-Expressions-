Q1 :- What is a Calculated Field and why is it important in business dashboards?
- Answer :- A Calculated Field in Tableau is a new field created using a formula based on existing data. It allows users to perform calculations, create new metrics, and analyze data without changing the original dataset.

Calculated Fields are important because they help convert raw data into meaningful business information, making reports and dashboards more useful for decision-making.


Importance of Calculated Fields in Business Dashboards
Create new business metrics such as Profit Margin, Discount Amount, and Total Revenue.
Perform calculations without modifying the original data.
Simplify complex data analysis.
Improve the accuracy and flexibility of reports.
Help management make faster and better business decisions.
Example

If a company wants to calculate Profit Margin, a Calculated Field can be created using the formula:

Profit Margin = (Profit ÷ Sales) × 100

The calculated value can then be used in charts and dashboards to evaluate product or regional performance.

Q2 :-  Why can’t parameters change visuals directly in Tableau?

- Answer :- A Parameter in Tableau is a user-defined input that stores a single value, such as a number, text, or date. By itself, a parameter cannot change a visualization directly because it is not connected to the data.

A parameter only stores a value. To affect a chart or dashboard, it must be used in a Calculated Field, Filter, or Reference Line. These connect the parameter to the data and allow the visualization to change based on the selected value.


- Why Parameters Cannot Change Visuals Directly

A parameter only stores one selected value.

It does not automatically filter or modify the dataset.

It must be linked to a Calculated Field or Filter to control a chart.

Without this connection, changing the parameter has no effect on the visualization.

Example

Suppose a parameter allows the user to choose Sales or Profit.

If the parameter is created alone, the chart will not change.
If it is linked to a Calculated Field, Tableau will display either Sales or Profit based on the user's selection.


 Q3 :- When should EXCLUDE LOD be used?

- Answer  :- EXCLUDE LOD (Level of Detail) is used when you want to remove one or more dimensions from a calculation while keeping those dimensions in the visualization. It helps calculate values at a higher level of detail than what is currently displayed in the view.


- When Should EXCLUDE LOD Be Used?

When you want to ignore a specific dimension in a calculation.

When you need totals or averages without considering one or more fields.

When comparing an individual value with an overall category or regional value.

When creating summary calculations while keeping detailed data visible.

Q4 :- Difference between FIXED and INCLUDE LOD expressions?

- Answer

FIXED and INCLUDE are both Level of Detail (LOD) expressions in Tableau, but they calculate data at different levels.

FIXED LOD

A FIXED LOD expression calculates values at a specified level of detail, regardless of the dimensions present in the view. It ignores the dimensions in the visualization unless they are included in the FIXED expression.

Example:

{ FIXED [Region] : SUM([Sales]) }

This calculates the total sales for each Region, even if the view displays data by Category or Product.



INCLUDE LOD

An INCLUDE LOD expression adds extra dimensions to the calculation in addition to the dimensions already present in the view. It calculates data at a more detailed level.

Example:

{ INCLUDE [Product Name] : SUM([Sales]) }

This includes Product Name in the calculation, even if it is not displayed in the view.


Q5 :- : Delivery Performance Classification (Calculated Field)

Business Need

Operations wants to track delivery efficiency.

Task

Classify orders as Fast or Delayed.

Condition

Delivery ≤ 30 min → Fast

Else → Delayed

- Answer

Step 1:

Open Tableau Desktop and connect to your dataset.

Step 2:

In the Data Pane, right-click anywhere and select Create Calculated Field.

Step 3:

Enter the name:

Delivery Status

Step 4:

Enter the following formula: IF [Delivery Time] <= 30 THEN
    "Fast"
ELSE
    "Delayed"
END


Step 5:

Click Apply, then click OK.

Step 6:

The new field Delivery Status will appear in the Data Pane.

Step 7:

Drag Delivery Status to Rows, Columns, Color, or Filters to analyze delivery performance.

Q7 :-  Top N Products Analysis (Parameter)

Business Need

The product team wants to focus on top-performing products.

Task

Create a Top N Products dynamic analysis.


- Answer

Step 1: Create a Parameter
In the Data Pane, right-click and select Create Parameter.

Step 2: Configure the Parameter
Name: Top N Products
Data Type: Integer
Current Value: 10
Allowable Values: Range (Minimum: 1, Maximum: 20, Step Size: 1)

Click OK  

Step 3: Show the Parameter
Right-click Top N Products.
Select Show Parameter.

Step 4: Create a Calculated Field

Right-click the Data Pane → Create Calculated Field.

Name: Top N Filter

Enter the formula: RANK(SUM([Sales])) <= [Top N Products]

Step 5: Build the Visualization
Drag Product Name → Rows
Drag Sales → Columns
Sort the products in Descending Order by SUM(Sales).


Step 6: Apply the Filter
Drag Top N Filter to the Filters shelf.
Select True.

Now the chart displays only the Top N Products based on the number selected in the parameter.

Q8 :- Customer/Product Detail Impact (INCLUDE LOD)

Business Need

The sales team wants average sales considering product-level detail, even when product is not shown.

Task

Create a calculation that finds the average sales at the product level, but allows the result to be used at a
higher level (such as Category or Region). Use INCLUDE LOD.  

- Answer

Step 1:

Open Tableau Desktop and connect to the dataset.

Step 2:

In the Data Pane, right-click and select Create Calculated Field.

Step 3:

Enter the name:

Average Sales per Product

Step 4:

Enter the following INCLUDE LOD formula: { INCLUDE [Product Name] : AVG([Sales]) }

Step 5:

Click Apply and then OK.

Step 6:

Create a visualization.

Drag Category (or Region) → Rows
Drag Average Sales per Product → Columns

Step 7 (Optional):

Sort the results in descending order to compare performance.
